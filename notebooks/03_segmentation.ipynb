```python
import pandas as pd
import numpy as np

clv = pd.read_csv('../data/processed/clv_scores.csv')
print(clv.shape)
print(clv.head())
```

```python
clv['Segment'] = pd.qcut(
    clv['CLV'],
    q=4,
    labels=['Lost', 'At Risk', 'Loyal', 'Champion']
)

print(clv['Segment'].value_counts())
print(clv.groupby('Segment')['CLV'].agg(['min', 'max', 'mean']).round(2))
```

```python
rfm = pd.read_csv('../data/processed/rfm_table.csv')
final = clv.merge(rfm, on='Customer ID', how='left')
print(final.shape)
print(final.head())
final.to_csv('../data/processed/final_customer_segments.csv', index=False)
print("Saved")
```
